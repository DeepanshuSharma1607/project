```python
def create_features(X):
    import numpy as np

    X = X.copy()

    X['bp_ratio'] = X['systolic_bp'] / (X['diastolic_bp'] + 1)
    X['activity_per_age'] = X['physical_activity_minutes_per_week'] / (X['age'] + 1)

    X['obese'] = (X['bmi'] > 30).astype(int)
    X['high_bp'] = (X['systolic_bp'] > 140).astype(int)
    X['low_activity'] = (X['physical_activity_minutes_per_week'] < 150).astype(int)
    X['poor_sleep'] = (X['sleep_hours_per_day'] < 6).astype(int)

    X['high_risk'] = (
        (X['obese']) &
        (X['high_bp']) &
        (X['low_activity'])
    ).astype(int)

    X['chol_ratio'] = X['ldl_cholesterol'] / (X['hdl_cholesterol'] + 1)
    X['total_hdl_ratio'] = X['cholesterol_total'] / (X['hdl_cholesterol'] + 1)

    # X['genetic_risk'] = X['family_history_diabetes'] * X['bmi']

    X['bmi_age'] = X['bmi'] * X['age']

    X['bp_chol_product'] = X['systolic_bp'] * X['cholesterol_total']

    X['lifestyle_score'] = (
        X['smoking_status_Current']*3 +
        X['low_activity']*2 +
        X['poor_sleep'] +
        X['obese']*2
    )

    X['old_age'] = (X['age'] > 50).astype(int)

    X['bp_genetic'] = X['systolic_bp'] * X['family_history_diabetes']
    X['bmi_genetic'] = X['bmi'] * X['family_history_diabetes']

    X['metabolic_score'] = (
    X['bmi'] +
    X['systolic_bp'] +
    X['cholesterol_total'] +
    X['triglycerides']
    )
    X['lifestyle_risk'] = (
    X['smoking_status_Current']*3 +
    X['low_activity']*2 +
    X['poor_sleep'] +
    X['screen_time_hours_per_day']
    )
    X['body_stress'] = (
    X['heart_rate'] +
    X['systolic_bp'] +
    X['bmi']
    )
    X['lipid_risk'] = (
    X['ldl_cholesterol'] +
    X['triglycerides'] -
    X['hdl_cholesterol']
)

    return X
```

```python
x_train = create_features(x_train)
x_cv = create_features(x_cv)
x_test = create_features(x_test)
```

```python
from xgboost import XGBClassifier
model_xgb = model_xgb = XGBClassifier(
    n_estimators=600,
    max_depth=7,
    learning_rate=0.03,
    subsample=0.8,
    colsample_bytree=0.8,
    gamma=1
)
model_xgb.fit(x_train,y_train)
```
